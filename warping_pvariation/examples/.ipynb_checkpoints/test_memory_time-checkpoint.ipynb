```python
import numpy as np
import torch
import math
from importlib import reload
import matplotlib.pyplot as plt
import seaborn
import pandas as pd

import pybnb
from transformers import *
from fbm import FBM
from pympler.classtracker import ClassTracker

import signatory 
import iisignature
from visualization_BnB_warped_pvar import BnBWarping
```

## Time complexity and Memory consumption as function of # of path evaluations

```python
p = 2.35
depth = np.floor(p).astype(int)
norm = 'l1'

time_complexity = []
memory = []
lengths = np.arange(3, 9)

for N in lengths:

    idx = np.linspace(0, 3*np.pi, N)
    x = np.sin(idx)
    y = np.cos(idx)

    x = np.array(AddTime().fit_transform([x]))
    y = np.array(AddTime().fit_transform([y]))
    
    x = signatory.Path(torch.tensor(x, dtype=torch.double), 2, basepoint=True)
    y = signatory.Path(torch.tensor(y, dtype=torch.double), 2, basepoint=True) 

    tracker = ClassTracker()
    problem = BnBWarping(x=x, y=y, p=p, width=2, norm=norm, root_node=(1,1), signator=True,
                         pvar_dist_mem=None, pvar_mem_org=None, initial_time=None, 
                         use_bound1=True, use_bound2=True, use_bound3=True)
    tracker.track_object(problem)
    tracker.create_snapshot()
    results = pybnb.Solver().solve(problem, log=None, queue_strategy='depth')
    tracker.create_snapshot()
    
    memory.append(tracker.snapshots[1].tracked_total/1000)
    time_complexity.append(results.wall_time)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15,6))
ax[0].plot(lengths, time_complexity, 'b--')
ax[1].plot(lengths, memory, 'r--')
ax[0].set_xlabel('length of 2 input paths', fontsize=12)
ax[1].set_xlabel('length of 2 input paths', fontsize=12)
ax[0].set_ylabel('secs from start', fontsize=12)
ax[1].set_ylabel('MB', fontsize=12)
ax[0].set_title('Time complexity', fontsize=12)
ax[1].set_title('Memory consumption', fontsize=12)
plt.show()
```

# Hyperparameter tuning (mainly p & boundary condition)

```python
# hyper-parameters
pvars = np.linspace(2., 2.99, 5)

# values to be stored
final_times = {}
final_nodes = {}
final_wpv = {}

# input paths specification
n_points = 5
norm = 'l1'

for p in pvars:
    
    depth = np.floor(p).astype(int)
    
    x = np.sin(np.linspace(0, 2*np.pi, n_points+2))
    y = np.cos(np.linspace(0, 3*np.pi, n_points))
    x = np.array(AddTime().fit_transform([x]))
    y = np.array(AddTime().fit_transform([y]))
    x = signatory.Path(torch.tensor(x, dtype=torch.double), 2, basepoint=True)
    y = signatory.Path(torch.tensor(y, dtype=torch.double), 2, basepoint=True) 
            
    # branch-and-bound
    problem = BnBWarping(x=x, y=y, p=p, width=2, norm=norm, root_node=(1,1), signator=True,
                         pvar_dist_mem=None, pvar_mem_org=None, initial_time=None, 
                         use_bound1=True, use_bound2=True, use_bound3=False)
    
    results = pybnb.Solver().solve(problem, log=None, queue_strategy='depth')

    final_times[f'p: {p}'] = results.wall_time
    final_nodes[f'p: {p}'] = results.nodes
    final_wpv[f'p: {p}'] = results.objective
```

```python
df_times = pd.Series(final_times)
df_nodes = pd.Series(final_nodes)
df_wpv = pd.Series(final_wpv)
```

```python
df = pd.DataFrame([df_times, df_nodes, df_wpv], index=['Algorithmic time', 'Processed nodes', 'warped p-var']).T
```

```python
plt.figure(figsize=(8, 6))
seaborn.heatmap(df, annot=True, cmap="YlGnBu", fmt='.1f')
plt.show()
```
