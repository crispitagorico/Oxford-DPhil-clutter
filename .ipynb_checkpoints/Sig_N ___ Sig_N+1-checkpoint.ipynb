```python
import tjl_dense_numpy_tensor
import numpy as np
import matplotlib.pyplot as plt
import tosig
```

```python
dim = 3
path = np.random.rand(100, dim)
```

```python
N = 4
```

```python
xxx = tosig.sigdim(dim, N) + 1
```

```python
sig_tensor = tjl_dense_numpy_tensor.stream2sigtensor(path, N)
```

```python
sig_tensor_plus1 = tjl_dense_numpy_tensor.stream2sigtensor(path, N+1)
```

```python
sig_tensor_logexp = tjl_dense_numpy_tensor.tensor_exp(tjl_dense_numpy_tensor.tensor_log(sig_tensor, N+1), N+1)
```

```python
diff = np.abs(sig_tensor_plus1 - sig_tensor_logexp)
```

```python
plt.figure(figsize=(15, 10))
plt.plot(diff)
plt.xlabel('TensorAlgebra elements', size=15)
plt.title('Absolute Difference between Sig_{N+1} and exp^{N+1}(log^N(Sig_N))', size=15)
plt.ylabel('||Sig_{N+1} - exp^{N+1}(log^N(Sig_N))||', size=15)
plt.axvline(xxx, color='red', lw=2, label='tensor element # {}'.format(xxx))
plt.legend()
plt.show()
```
